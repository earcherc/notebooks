```python
# https://huggingface.co/datasets/xsum
from datasets import load_dataset

dataset = load_dataset("xsum")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# print(dataset)
# There are three dicts within the dataset - train, validation, test
# Each dataset contains two keys: 'features' and 'num_rows'
# 'features' contains three columns: 'document', 'summary' and 'id'

df = pd.DataFrame(dataset['train'])
# df.head()
df.head()
```

```python
# Get some basic info on each row of documents and summaries

df['doc_char_len'] = df['document'].apply(len)
df['doc_token_len'] = df['document'].apply(lambda x: len(x.split()))
df['doc_sentence_len'] = df['document'].apply(lambda x: len(x.split('. ')))
df['doc_sentence_len2'] = df['document'].apply(lambda x: len(x.split('.')))

df['summary_char_len'] = df['document'].apply(len)
df['summary_token_len'] = df['document'].apply(lambda x: len(x.split()))
df['summary_sentence_len'] = df['document'].apply(lambda x: len(x.split('. ')))
df['summary_sentence_len2'] = df['document'].apply(lambda x: len(x.split('.')))
```

```python
# Get some basic stats on the data we've processed

# Calculate mean and standard deviation for document-related features
mean_doc_char_len = df['doc_char_len'].mean()
std_doc_char_len = df['doc_char_len'].std()

mean_doc_token_len = df['doc_token_len'].mean()
std_doc_token_len = df['doc_token_len'].std()

mean_doc_sentence_len = df['doc_sentence_len'].mean()
std_doc_sentence_len = df['doc_sentence_len'].std()

mean_doc_sentence_len2 = df['doc_sentence_len2'].mean()
std_doc_sentence_len2 = df['doc_sentence_len2'].std()

# Calculate mean and standard deviation for summary-related features
mean_summary_char_len = df['summary_char_len'].mean()
std_summary_char_len = df['summary_char_len'].std()

mean_summary_token_len = df['summary_token_len'].mean()
std_summary_token_len = df['summary_token_len'].std()

mean_summary_sentence_len = df['summary_sentence_len'].mean()
std_summary_sentence_len = df['summary_sentence_len'].std()

mean_summary_sentence_len2 = df['summary_sentence_len2'].mean()
std_summary_sentence_len2 = df['summary_sentence_len2'].std()

# Create a DataFrame to display these statistics
data = {
    'Feature': ['Document', 'Summary'],
    'Mean Char Len': [mean_doc_char_len, mean_summary_char_len],
    'Std Dev Char Len': [std_doc_char_len, std_summary_char_len],
    'Mean Token Len': [mean_doc_token_len, mean_summary_token_len],
    'Std Dev Token Len': [std_doc_token_len, std_summary_token_len],
    'Mean Sentence Len': [mean_doc_sentence_len, mean_summary_sentence_len],
    'Std Dev Sentence Len': [std_doc_sentence_len, std_summary_sentence_len],
    'Mean Sentence Len 2': [mean_doc_sentence_len2, mean_summary_sentence_len2],
    'Std Dev Sentence Len 2': [std_doc_sentence_len2, std_summary_sentence_len2],
}

stats_df = pd.DataFrame(data).set_index('Feature')
stats_df
```
