```python
import spacy
from spacy import displacy
spacy.cli.download("en_core_web_sm")
```

```python
from collections import Counter
nlp = spacy.load("en_core_web_sm")
import pandas as pd

sentence = 'It has also arisen in criminal justice, healthcare, and hiring, compounding existing racial, economic, and gender biases.'
doc = nlp(sentence)
displacy.render(doc, style='dep', jupyter=True)
displacy.render(doc, style='ent', jupyter=True)
tokens = [token.text for token in doc]
pos_tags = [(token.text, token.pos_) for token in doc]
lemmas = [(token.text, token.lemma_) for token in doc]
sentences = [sent.text for sent in doc.sents]
vectors = [(token.text, token.vector) for token in doc]
print(f'~~~~~~~Tokens~~~~~~\n {tokens}' )
# print(f'~~~~~~~Pos Tags~~~~ {pos_tags}')
# print(f'~~~~~~~Lemmas~~~~~~ {lemmas}')
# print(f'~~~~~~~Sentences~~~ {sentences}')
# print(f'~~~~~~~Vectors~~~~~ {vectors}')

bag_of_words = Counter(tokens)
# print(f'~~~~~~BoW~~~~~~ {bag_of_words}')
print(f'~~~~~~BoW~~~~~~\n {bag_of_words.most_common(3)}')

counts = pd.Series(dict(bag_of_words.most_common()))
print(f'~~~~~~TF~~~~~~\n {counts}')

print(f'~~~~~~Relative TF~~~~~~\n{counts / counts.sum()}')
```
